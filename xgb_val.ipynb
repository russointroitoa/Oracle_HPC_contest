```python
import pandas as pd
```

```python
s = pd.read_csv('lgb_sub8_scores.csv')
s
```

```python
group_queried = s[['queried_record_id', 'rec_pred']].groupby('queried_record_id').apply(lambda x: list(x['rec_pred']))
s = pd.DataFrame(group_queried).reset_index().rename(columns={0 : 'rec_pred'})
```

```python
def reorder_preds(preds):
    sorted_list = []
    for i in range(len(preds)):
        l = sorted(preds[i], key=lambda t: t[1], reverse=True)
        sorted_list.append(l)
    return sorted_list
```

```python
from tqdm.auto import tqdm
```

```python
rec_pred_corr = [[eval(x) for x in t] for t in tqdm(s.rec_pred)]
s['rec_pred'] = rec_pred_corr
```

```python
s['reordered_preds'] = reorder_preds(s.rec_pred.values)
```

```python
s
```

```python
s['first_pred'] = [x[0] for x in s.reordered_preds]
s
```

```python
first = []
for (q, f) in zip(s.queried_record_id, s.first_pred):
    if f[1] < 0:
        first.append(q)
    else:
        first.append(f[0])
first
```

```python
s['predicted_record_id'] = first
```

```python
s[['queried_record_id', 'predicted_record_id']].to_csv("greater_than_0_lgb.csv", index=False)
```
