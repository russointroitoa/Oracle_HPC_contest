```python
import pandas as pd
from tqdm.auto import tqdm
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 20)
```

```python
df_train = pd.read_csv("../dataset/original/entity-resolution_advanced-topics-training_data.csv")
df_train.head(10)
```

```python
df_train = df_train[['record_id','name','linked_id']]
df_train.head(10)
```

```python
# convert to lowercase
df_train.name = df_train.name.astype(str)
df_train.name = df_train.name.str.lower()
df_train.head(10)
```

```python
# remove all punctuation
import string
df_train.name = df_train.name.str.translate(str.maketrans('', '', string.punctuation))
df_train.head(10)
```

```python
# remove accented characters
import unidecode
no_accents = []
for s in df_train.name:
    no_accents.append(unidecode.unidecode(s))
df_train.name = no_accents
df_train.head(10)
```

```python
# create columns for all the letters
columns = ['record_id','name','a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
name_matrix = pd.DataFrame(columns=columns)
name_matrix
```

```python
name_matrix.record_id = df_train.record_id.copy()
name_matrix.name = df_train.name.copy()
name_matrix.head(10)
```

```python
for l in tqdm(['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']):
    new_col = []
    for (i,n) in zip(name_matrix.index, name_matrix.name):
        new_col.append(n.count(l))
    name_matrix[l] = new_col
name_matrix.head(10)
```

```python
import string
import unidecode
def create_name_letters_matrix(df):
    df = df[['record_id','name']]
    df.name = df.name.astype(str) # convert to string
    df.name = df.name.str.lower() # lowercase
    df.name = df.name.str.translate(str.maketrans('', '', string.punctuation)) # remove punctuation
    # remove accented letters
    no_accents = []
    for s in df.name:
        no_accents.append(unidecode.unidecode(s))
    df.name = no_accents
    # create return matrix
    columns = ['record_id','name','a','b','c','d','e','f','g','h','i','j','k','l',
               'm','n','o','p','q','r','s','t','u','v','w','x','y','z']
    name_letters_matrix = pd.DataFrame(columns=columns)
    name_letters_matrix.record_id = df.record_id.copy()
    name_letters_matrix.name = df.name.copy()
    # count occurence of each letter and add the columns to the return df
    for l in tqdm(['a','b','c','d','e','f','g','h','i','j','k','l','m','n',
                   'o','p','q','r','s','t','u','v','w','x','y','z']):
        new_col = []
        for (i,n) in zip(name_letters_matrix.index, name_letters_matrix.name):
            new_col.append(n.count(l))
        name_letters_matrix[l] = new_col
    return name_letters_matrix
```

```python
create_name_letters_matrix(df_train)
```
