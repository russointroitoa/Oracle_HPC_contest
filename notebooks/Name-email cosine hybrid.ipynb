```python
import pandas as pd
from scipy import *
from scipy.sparse import *
```

```python
from tqdm.auto import tqdm
```

```python
sim_name = load_npz('../similarity_cosine_complete_name.npz')
sim_email = load_npz('../similarity_cosine_complete_email.npz')
hybrid = sim_name + 0.3*sim_email
```

```python
len(sim_name[0].nonzero()[1]) #ERRORE : più dati che indici non zero
```

```python
sim = hybrid
```

```python
df_train = pd.read_csv('../dataset/original/train.csv', escapechar="\\")
df_test = pd.read_csv('../dataset/original/test.csv', escapechar="\\")
```

```python
df_train = df_train.sort_values(by=['record_id']).reset_index(drop=True)
df_test = df_test.sort_values(by=['record_id']).reset_index(drop=True)
```

```python
linid = []
for x in tqdm(range(df_test.shape[0])):
    if x == 0:
        #Dobbiamo considerare separatamente il primo elemento perché c'è un errore nella similarità: in questo caso
        # vengono considerati 101 dati invece di 100, mentre gli indici corrispondenti sono 100. Di conseguenza
        # gli indici che vengono tirati fuori sono sballati 
        linid.append(df_train.loc[sim[x].nonzero()[1][sim[x].data[1:].argsort()[::-1][:100]],:].drop_duplicates(['linked_id'])['linked_id'].values[:10])
    else:
        linid.append(df_train.loc[sim[x].nonzero()[1][sim[x].data.argsort()[::-1][:100]],:].drop_duplicates(['linked_id'])['linked_id'].values[:10])
```

```python
sub = pd.DataFrame()
sub['queried_record_id'] = df_test.record_id
```

```python
sub['predicted_record_id'] = linid
sub
```

```python
strings = []
for t in tqdm(sub.predicted_record_id):
    strings.append(' '.join([str(x) for x in t]))
```

```python
sub['predicted_record_id'] = strings
sub
```

```python
sub.to_csv('sub_fixed_2.csv', index=False)
```
