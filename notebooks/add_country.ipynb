```python
import pandas as pd
from tqdm.auto import tqdm
import numpy as np
```

```python
df_train = pd.read_csv('../dataset/original/entity-resolution_advanced-topics-training_data.csv')
```

```python
df_train.phone = df_train.phone.fillna(-1).astype(np.int64)
```

```python
no_dupl = len(set(df_train.name))
no_dupl
```

```python
df_train[df_train.name == 'ALDIA ENTERPRISES LTD.']
```

```python
from iso3166 import countries
```

```python
df_train.email = df_train.email.astype(str)
```

```python
new_col = []
for m in tqdm(df_train.email):
    if m != 'nan':
        new_col.append(m.split('.')[-1])
    else:
        new_col.append('nan')
```

```python
df_train['mail_country'] = new_col
df_train.head(100)
```

```python
new_col = []
for m in tqdm(df_train.mail_country):
    if m != 'nan':
        try:
            new_col.append(countries.get(m).alpha2)
        except KeyError:
            new_col.append('nan')
    else:
        new_col.append('nan')
```

```python
df_train['country_from_mail'] = new_col
df_train.drop(['mail_country'],1)
```

```python
import phonenumbers
n = phonenumbers.parse("+65013464145", None)
n
```

```python
phonenumbers.region_code_for_country_code(n.country_code)
```

```python
new_col = []
for pn in tqdm(df_train.phone):
    if pn != 'nan':
        try:
            new_col.append(phonenumbers.region_code_for_country_code(phonenumbers.parse('+'+str(pn), None).country_code))
        except:
            new_col.append('not found')

    else:
        new_col.append('nan')
```

```python
df_train['country_from_phone'] = new_col
df_train
```
