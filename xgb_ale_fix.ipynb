```python
import pandas as pd
pd.set_option('display.max_rows', 500)
```

```python
xgb_pred = pd.read_csv('xgb_sub8_scores.csv')
xgb_pred.head(1000)
```

```python
df_train = pd.read_csv('dataset/original/train.csv', escapechar='\\')
```

```python
df_train.linked_id = df_train.linked_id.astype(str)
tr_lid = list(set(df_train.linked_id))
test_lid = list(set([str(x).split('-')[0] for x in xgb_pred.queried_record_id]))
```

```python
only_test = set(test_lid)-set(tr_lid)
```

```python
len(only_test)
```

```python
xgb_pred['lin_id'] = [str(x).split('-')[0] for x in xgb_pred.queried_record_id]
xgb_pred.head()
```

```python
xgb_true = xgb_pred[~xgb_pred.lin_id.isin(only_test)]
```

```python
group_queried = xgb_true[['queried_record_id', 'rec_pred']].groupby('queried_record_id').apply(lambda x: list(x['rec_pred']))
xgb_true = pd.DataFrame(group_queried).reset_index().rename(columns={0 : 'rec_pred'})
xgb_true
```

```python
from tqdm.auto import tqdm
```

```python
rec_pred_corr = [[eval(x) for x in t] for t in tqdm(xgb_true.rec_pred)]
```

```python
xgb_true.rec_pred = rec_pred_corr
```

```python
def reorder_preds(preds):
    sorted_list = []
    for i in tqdm(range(len(preds))):
        l = sorted(preds[i], key=lambda t: t[1], reverse=True)
        sorted_list.append(l)
    return sorted_list
```

```python
xgb_true['ordered_preds'] = reorder_preds(xgb_true.rec_pred.values)
```

```python
xgb_true
```

```python
negative_first_pred = []
for (q, o) in zip(xgb_true.queried_record_id, xgb_true.ordered_preds):
    if o[0][1] < 2:
        negative_first_pred.append((q, o[0]))
negative_first_pred
```

```python
len(negative_first_pred)
```
